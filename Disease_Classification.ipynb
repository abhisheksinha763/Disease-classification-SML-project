# ***Objective 2) Applying PCA to see how classification accuracy decreases as dimension is reduced***

Here for each algorithms we have printed the accuracy of the algorithms by taking features from 1 to 20 and then ploted a graph of accuracy v/s features by creating a function which takes the model as input.
As the data is very simple in each case we can see that we are getting accuracy greater than 95% with features greater than equal to 5.

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.decomposition import PCA
```

```python
def fn(model):
  accuracy_model = []
  features = []
  for i in range(131):
      pca = PCA(n_components=(i+1))
      X_train_pca = pca.fit_transform(X_train)
      X_test_pca = pca.transform(X_test)
      model.fit(X_train_pca, Y_train)
      accuracy_model.append(accuracy_score(Y_test, model.predict(X_test_pca))*100)
      features.append(i+1)
      if(i < 20 ):
        print("Accuracy on the testing data with ",i+1," feature is ",accuracy_score(Y_test, model.predict(X_test_pca))*100,'%')
  return accuracy_model,features
```

```python
#(1)
accuracy_lr,features = fn(clf_lr)
plt.plot(features, accuracy_lr, label='Logistic Regression')
plt.xlabel("features")
plt.ylabel("accuracy")
plt.title("features vs accuracy")

plt.show()
```

```python
#(2)
accuracy_svm,features = fn(clf_svm)
plt.plot(features, accuracy_svm, label='Support Vector Machines')
plt.xlabel("features")
plt.ylabel("accuracy")
plt.title("features vs accuracy")

plt.show()
```

```python
#(3)
accuracy_knn,features = fn(clf_knn)
plt.plot(features, accuracy_knn, label='K Nearest Neighbour')
plt.xlabel("features")
plt.ylabel("accuracy")
plt.title("features vs accuracy")

plt.show()
```

```python
#(4)
accuracy_Dt,features = fn(clf_Dt)
plt.plot(features, accuracy_Dt, label='Decision Tree')
plt.xlabel("features")
plt.ylabel("accuracy")
plt.title("features vs accuracy")

plt.show()
```

```python
#(5)
accuracy_lda,features = fn(clf_lda)
plt.plot(features, accuracy_lda, label='Linear Discriminant Analysis')
plt.xlabel("features")
plt.ylabel("accuracy")
plt.title("features vs accuracy")


plt.show()
```

```python
#(6)
accuracy_gnb,features = fn(clf_gnb)
plt.plot(features, accuracy_gnb, label='Naive Bayes')
plt.xlabel("features")
plt.ylabel("accuracy")
plt.title("features vs accuracy")

plt.legend()

plt.show()
```

## ***Objective 3) Applying Random Forest algorithm and tuning the hyper-parameters to get the best classification accuracy.***

```python
from sklearn.ensemble import RandomForestClassifier
```

### Tuning the Hyperparameters -: we can tune the Hyperparameters using GridSearchCV which takes different parameters as key and different values of the paramters (in a array) as value in a dictionary and gives the best parameters on testing on validation data

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf_Rf = RandomForestClassifier()
grid = {
        'max_depth': [10,50,100],
        'min_samples_leaf': [1, 3, 6],
        'criterion' : ['entropy','gini', 'log_loss'],
        'min_samples_split': [1, 3, 5],
        'max_features': [2, 4,6,],
        'n_estimators': [100,500,1000]
}
Rf_with_tuned_parameters =  GridSearchCV(clf_Rf, grid).fit(X_train, Y_train)
```

### To see what are the best parameters which we get using gridsearchcv we can use best_estimator_ which will give the best parameters among parameters which we have provided it and cv_results_ will give score on all the parameters.

```python
Rf_with_tuned_parameters.best_estimator_
```

clearly we can see out of all values provided the best values for max_depth = 10, max_features = 2, min_samples_split = 3 and criterion = entropy.

Below code can be used to get the split score for all possible combination of different parameters that we have provided.

```python
#Rf_with_tuned_parameters.cv_results_
```

```python
print("Accuracy on the training data is ",accuracy_score(Y_train, Rf_with_tuned_parameters.predict(X_train)))
print("Accuracy value on the testing data is ",accuracy_score(Y_test, Rf_with_tuned_parameters.predict(X_test)))
```
